#### logisic sigmoid: 
$ σ(z) = \frac {1}{1+e^{-z}} = \frac{e^z}{e^z+1} $

#### Tương tự là hàm tanh:

$ f(z) = tanh(z) = 2σ(2z) - 1 = \frac{e^z - e^{-z}}{e^z + e^{-z}} = \frac{e^{2z} - 1}{e^{2z} + 1} = \frac {1-e^{-2z}}{1 + e^{-2z}}$

#### Dạng tổng quát
$ f(x) = σ(ax + b), f(x) = tanh (ax + b)$
